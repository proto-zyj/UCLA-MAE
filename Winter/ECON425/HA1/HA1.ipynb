```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### ReLU function:

```python
ax = plt.subplot(111)
x = np.arange(-10,10)
ReLU = [max(i,0) for i in x]
line = plt.plot(x,ReLU)
plt.ylim(-2,10)
plt.show()
```

### Leaky ReLU function

```python
ax = plt.subplot(111)
x = np.arange(-10,10)
leakyReLU = [max(0.1 * i,i) for i in x]
line = plt.plot(x,leakyReLU)
plt.ylim(-2,10)
plt.show()
```

### Tanh(x)

```python
ax = plt.subplot(111)
x = np.arange(-5,5)
tanh = np.tanh(x)
line = plt.plot(x,tanh)
plt.show()
```

### Straight line

```python
ax = plt.subplot(111)
x = np.arange(-5,5)
y = 30 + 0.5 * x
line = plt.plot(x,y)
plt.show()
```

### Quadratic function

```python
ax = plt.subplot(111)
x = np.arange(20,31)
y = (x - 25) ** 2 + 20
line = plt.plot(x,y)
plt.show()
```

###  Log function

```python
ax = plt.subplot(111)
x = np.arange(1,10)
y = - np.log(x)
line = plt.plot(x,y)
plt.show()
```

```python
ax = plt.subplot(111)
x = np.arange(-5,1)
y = - np.log(1 - x)
line = plt.plot(x,y)
plt.show()
```

### Sigmoid Function

```python
ax = plt.subplot(111)
x = np.arange(-5,5)
y = 1 / (1 + np.exp(-x))
line = plt.plot(x,y)
plt.show()
```
